# Statistical Analysis

```python
# for the Observed Length (m) :

print("Mean:", df['Observed Length (m)'].mean())
print("Median:", df['Observed Length (m)'].median())
print("Mode:", df['Observed Length (m)'].mode()[0])
print("Standard Deviation:", df['Observed Length (m)'].std())
print("Minimum:", df['Observed Length (m)'].min())
print("Maximum:", df['Observed Length (m)'].max())
```

NOTE:
1. on an average the length of crocodiles is 2.41meters.
2. The most common length is 2.26 meters.

3. The minimum length of the crocodile is 0.14 meters, which may be when the crocodile is just born
4. The maximum length of the crocodile is 6.12 meters.

```python
# for the Observed Weight (kg):
print("Mean:", df['Observed Weight (kg)'].mean())
print("Median:", df['Observed Weight (kg)'].median())
print("Mode:", df['Observed Weight (kg)'].mode()[0])
print("Standard Deviation:", df['Observed Weight (kg)'].std())
print("Minimum:", df['Observed Weight (kg)'].min())
print("Maximum:", df['Observed Weight (kg)'].max())
```

NOTE:
1. On  an average, the crocodiles weigh about 156 kg.
2. The median value is 100.6 kg, which is less than the mean. so it means that the the distribution is right-skewed.
3. The most frequently weight of the crocodile is 11.3 kg.
4. The minimum weight of the  crocodile is  4.4 kg, probably a newborn or may be very young crocodile.
5. The maximum weight of the crocodile is 1139.7 kg
